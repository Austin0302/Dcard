# 4.爬取多篇

```python
def Crawl(ID):
    link = 'https://www.dcard.tw/_api/posts/' + str(ID)
    requ = requests.get(link, headers = {
    "user-agent": ua.random
  })
    rejs = requ.json()
    return(pd.DataFrame(
        data=
        [{'id':rejs['id'],
          '標題':rejs['title'],
          '摘要':rejs['excerpt'],
          '校名匿名':rejs['anonymousSchool'],
          '匿名':rejs['anonymousDepartment'],
          '發文時間':rejs['createdAt'],
          '更新時間':rejs['updatedAt'],
          '回應數量':rejs['commentCount'],
          '愛心數量':rejs['likeCount'],
          '標註':rejs['topics'],
          '分類':rejs['forumName'],
          '性別':rejs['gender'],
          '心情細項':rejs['reactions']}],
        columns=['id','標題','摘要','校名匿名','匿名','發文時間','更新時間','回應數量','愛心數量','標註','分類','性別','心情細項']))
```

```python
Crawl(235893073)
```

```python
url = 'https://www.dcard.tw/_api/posts?popular=true&limit=10'
resq = requests.get(url, headers = {
    "user-agent": ua.random
  })
rejs = resq.json()
df = pd.DataFrame()
for i in range(len(rejs)):
    df = df.append(Crawl(rejs[i]['id']),ignore_index=True)
print(df.shape)
df
```

```python
delay_choices = [3, 4, 5, 6, 7]
delay = random.choice(delay_choices)

for j in range(10):
    last = str(int(df.tail(1).id))
    url = 'https://www.dcard.tw/_api/posts?popular=true&limit=10&before=' + last
    resq = requests.get(url, headers = {
    "user-agent": ua.random
  })
    rejs = resq.json()
    time.sleep(delay)
    for i in range(len(rejs)):
        df = df.append(Crawl(rejs[i]['id']), ignore_index=True)
print(df.shape)
df
```
