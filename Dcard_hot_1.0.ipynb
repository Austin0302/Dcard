# 4.çˆ¬å–å¤šç¯‡

```python
def Crawl(ID):
    link = 'https://www.dcard.tw/_api/posts/' + str(ID)
    requ = requests.get(link, headers = {
    "user-agent": ua.random
  })
    rejs = requ.json()
    return(pd.DataFrame(
        data=
        [{'id':rejs['id'],
          'æ¨™é¡Œ':rejs['title'],
          'æ‘˜è¦':rejs['excerpt'],
          'æ ¡ååŒ¿å':rejs['anonymousSchool'],
          'åŒ¿å':rejs['anonymousDepartment'],
          'ç™¼æ–‡æ™‚é–“':rejs['createdAt'],
          'æ›´æ–°æ™‚é–“':rejs['updatedAt'],
          'å›æ‡‰æ•¸é‡':rejs['commentCount'],
          'æ„›å¿ƒæ•¸é‡':rejs['likeCount'],
          'æ¨™è¨»':rejs['topics'],
          'åˆ†é¡':rejs['forumName'],
          'æ€§åˆ¥':rejs['gender'],
          'å¿ƒæƒ…ç´°é …':rejs['reactions']}],
        columns=['id','æ¨™é¡Œ','æ‘˜è¦','æ ¡ååŒ¿å','åŒ¿å','ç™¼æ–‡æ™‚é–“','æ›´æ–°æ™‚é–“','å›æ‡‰æ•¸é‡','æ„›å¿ƒæ•¸é‡','æ¨™è¨»','åˆ†é¡','æ€§åˆ¥','å¿ƒæƒ…ç´°é …']))
```

```python
url = 'https://www.dcard.tw/_api/posts?popular=true&limit=10'
resq = requests.get(url, headers = {
    "user-agent": ua.random
  })
rejs = resq.json()
df = pd.DataFrame()
for i in range(len(rejs)):
    df = df.append(Crawl(rejs[i]['id']),ignore_index=True)
print(df.shape)
df
```

```python
delay_choices = [3, 4, 5, 6, 7]
delay = random.choice(delay_choices)

for j in range(10):
    last = str(int(df.tail(1).id))
    url = 'https://www.dcard.tw/_api/posts?popular=true&limit=10&before=' + last
    resq = requests.get(url, headers = {
    "user-agent": ua.random
  })
    rejs = resq.json()
    time.sleep(delay)
    for i in range(len(rejs)):
        df = df.append(Crawl(rejs[i]['id']), ignore_index=True)
print(df.shape)
df
```
